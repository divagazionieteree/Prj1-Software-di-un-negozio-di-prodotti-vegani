Software gestionale per negozio di prodotti vegani

import class and libraries

```python
import json as json
from datetime import datetime
```

Read .json files or create if there aren't in the project folder

```python
warehouse_json = "warehouse.json"
sales_json = "sales.json"

try:
  with open(warehouse_json, "r") as json_file:
    warehouse_dict = json.load(json_file)
except:
  open(warehouse_json, "x")
  warehouse_dict = []

try:
  with open(sales_json, "r") as json_file:
    sales_dict = json.load(json_file)
except:
  open(sales_json, "x")
  sales_dict = []
```

Collection of functions to be used in the project

```python
def aggiungi(product_name_f, quantity_f, purchase_price_f, selling_price_f, warehouse_dict_f):

    """
    Add a new product to warehouse with quantity, purchase prize and selling
    prize or update only the quantity if the product is still inside
    the warehouse
    """

    product_dict = {"product_name": product_name_f, "quantity": quantity_f, "purchase_prize": purchase_price_f, "sale_prize": selling_price_f}
    warehouse_dict_f.append(product_dict)

    print(f"AGGIUNTO {quantity_f} X {product_name_f}")

def elenca(warehouse_dict_f):

    """
    show all products in the warehouse
    """

    for item in warehouse_dict_f:
        print(item)
    #return print(wharehouse_dict, end="\n")

def vendita(product_name_f, quantity_f, sales_dict_f, warehouse_dict_f):

    """
    Register a sales operation
    """

    is_product = 0

    for i, val in enumerate(warehouse_dict_f):
      if warehouse_dict_f[i].get("product_name") == product_name_f:
        is_product += 1
        if warehouse_dict_f[i]["quantity"] >= quantity_f:
            current_date = str(datetime.now())
            cost = warehouse_dict_f[i]["purchase_prize"] * quantity_f
            billing = warehouse_dict_f[i]["sale_prize"] * quantity_f
            selling_dict = {"product_name": product_name_f, "quantity": quantity_f, "date": current_date, "cost": cost, "billing": billing }
            sales_dict_f.append(selling_dict)
            warehouse_dict_f[i]["quantity"] -= quantity_f
        else:
            print(f"non ci sono abbastanza prodotti")

    if is_product == 0:
      print("prodotto non presente")

def profitti(sales_dict_f):

    """
    Show Total Income and total Margin
    """

    total_cost = 0
    total_billing = 0

    for i, val in enumerate(sales_dict_f):
        total_cost = total_cost + sales_dict_f[i]["cost"]
        total_billing = total_billing + sales_dict_f[i]["billing"]

    total_margin = total_billing - total_cost
    print(f"Il fatturato totale è: {total_billing}")
    print(f"Il margine totale è: {total_margin}")

def aiuto():

    """
    Show all possible commands
    """

    str_aiuto = " I comandi disponibili sono i seguenti: \n aggiungi: aggiungi un prodotto al magazzino \n elenca: elenca i prodotto in magazzino \n vendita: registra una vendita effettuata \n profitti: mostra i profitti totali \n aiuto: mostra i possibili comandi \n chiudi: esci dal programma"
    return print(str_aiuto)

def chiudi(warehouse_json_f, sales_json_f, warehouse_dict_f, sales_dict_f):

    """
    Exit from the loop and write variables inside .json file to store
    them permanently
    """

    with open(warehouse_json_f, "w") as outfile:
        json.dump(warehouse_dict_f, outfile)

    with open(sales_json_f, "w") as outfile:
        json.dump(sales_dict_f, outfile)

    print("Grazie e Buona giornata")
```

Project to run

```python
cmd = None

while True:

  cmd = input("Inserisci un comando: ")

  if cmd=="vendita":
    # Register a selling operation

    inserting_continue = "si"
    while inserting_continue == "si":

      list_products = []
      for i, val in enumerate(warehouse_dict):
        list_products.append(warehouse_dict[i].get("product_name"))

      product_name = None
      while product_name == None or product_name not in list_products:
        try:
          product_name = input("Prodotto da acquistare: ")
        except:
          print("prodotto non presente")

      quantity = None
      while quantity is None or quantity <= 0:
        try:
          quantity = int(input("Quantità: "))
        except:
          print("dato non valido")

      vendita(product_name, quantity, sales_dict, warehouse_dict)
      inserting_continue = input("vuoi aggiungere un'altra vendita? 'si' / 'no' o qualsiasi altro comando ")

  elif cmd=="profitti":
    # Show Total Income and total Margin

    profitti(sales_dict)

  elif cmd=="aggiungi":
    # Add new products to warehouse

    inserting_continue = "si"

    while inserting_continue == "si":

      product_name = input("Nome del prodotto: ")

      quantity = None
      while quantity is None  or quantity <= 0:
        try:
          quantity = int(input("Quantità: "))
        except:
          print("dato non valido")

      is_product = 0

      for i, val in enumerate(warehouse_dict):
        if warehouse_dict[i]["product_name"] == product_name:
          is_product +=1
          prod_index = i

      if is_product > 0:
        warehouse_dict[prod_index]["quantity"] += quantity
        print(f"AGGIUNTO {quantity} X {product_name}")
      else:
        purchase_price = None
        while purchase_price is None or purchase_price <= 0:
          try:
            purchase_price = float(input("Prezzo di acquisto: "))
          except:
            print("dato non valido")

        selling_price = None
        while selling_price is None or selling_price <= 0:
          try:
            selling_price = float(input("Prezzo di vendita: "))
          except:
            print("dato non valido")

        aggiungi(product_name, quantity, purchase_price, selling_price, warehouse_dict)

      inserting_continue = input("vuoi aggiungere un'altro prodotto? 'si' / 'no' o qualsiasi altro comando ")

  elif cmd=="elenca":
    # Show all products in the warehouse

    elenca(warehouse_dict)

  elif cmd=="aiuto":
    # show all possible commands

    aiuto()

  elif cmd=="chiudi":
    # Close the application saving permanently warehouse and sales

    chiudi(warehouse_json, sales_json, warehouse_dict, sales_dict)
    break

  else:
    # Show help message

    print("comando non valido")
    aiuto()
```
